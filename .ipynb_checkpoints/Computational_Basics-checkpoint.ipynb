## Question 1

### a)

```python
primes = []
for i in range(2, 101):
    flag = True
    for j in range(2,i):
        if i % j == 0:
            flag = False
    if flag == True:
        primes.append(i)
print(primes)    
```

### b)

```python
form_primes = []
for i in range(1,11):
    odd_num = (2**i) - 1
    for num in primes:
        if odd_num == num:
            form_primes.append(odd_num)
print(form_primes)
```
