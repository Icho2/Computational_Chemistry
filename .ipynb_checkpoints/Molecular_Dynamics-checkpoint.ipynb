## Part A: Diatomic Molecules
### A1: The Lennard-Jones Potential
#### a)

```python
epsilon = 1.0
sigma = 1.0

x = np.linspace(0.99, 2.5)
potential = 4*epsilon*((sigma/x)**12 - (sigma/x)**6)

plt.plot(x, potential); plt.xlabel('r'); plt.ylabel("V(r)"); plt.title("Lennard-Jones Potential")
```
